```python
hypertension_DF.describe()
```

```python
#- Are there any duplicate rows?
hypertension_DF.duplicated().sum()
```

```python
# Calculate the mean median and mode for each column

for col in hypertension_DF.select_dtypes(include='number').columns:
    print(f"\nColumn: {col}")
    print(f"Mean: {hypertension_DF[col].mean()}")
    print(f"Median: {hypertension_DF[col].median()}")
    print(f"Mode: {hypertension_DF[col].mode().tolist()}")
```

```python
# Are there any missing or null values?

hypertension_DF.isnull().sum()
```

```python
#- Identify any other inconsistent data (e.g. someone seems to be taking an action before they are born.)


print("Ages over 120:")
print(hypertension_DF[hypertension_DF['Age'] > 120])

print("\nSystolic BP over 300:")
print(hypertension_DF[hypertension_DF['Systolic_BP'] > 300])

print("\nSleep Duration over 24 hours:")
print(hypertension_DF[hypertension_DF['Sleep_Duration'] > 12])

```
